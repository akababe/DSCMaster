```python
import plotly.figure_factory as ff
import plotly.offline as offline
import plotly.graph_objs as go
offline.init_notebook_mode()
import numpy as np
import pandas as pd

trace1 = go.Bar(
    x=hasil.shift_kerja,
    y=hasil.Omron ,
    name='Omron'
)
trace2 = go.Bar(
    x=hasil.shift_kerja,
    y=hasil.AB,
    name='AB'
)
trace3 = go.Line(
    x=hasil.shift_kerja,
    y=hasil.setpo,
    name='set'
)

data = [trace1,trace2,trace3]
layout = go.Layout(
    barmode='group',
    title = 'A'
)

fig = go.Figure(data=data, layout=layout)
offline.iplot(data, filename= "Beruang-Madu")
```
